```python
import pandas as pd
```

```python
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
```

```python
movies.head()
```

```python
ratings.head()
```

```python
data = pd.merge(ratings, movies, on = 'movieId')
```

```python
user_counts = data['userId'].value_counts()
```

```python
data = data[data['userId'].isin(user_counts[user_counts > 50].index)]
```

```python
movie_counts = data['title'].value_counts()
```

```python
data = data[data['title'].isin(movie_counts[movie_counts > 50].index)]
```

```python
user_movie = data.pivot_table(index='userId', columns='title', values='rating')
```

```python
user_movie = user_movie.fillna(0)
```

```python
user_movie = (user_movie >= 3).astype(bool)
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
frequent_items = fpgrowth(user_movie, min_support=0.02, use_colnames=True)
```
